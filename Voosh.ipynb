```python
import re
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set(rc={'figure.figsize':(12,12)})
sns.set_style('whitegrid')
```

```python
for i in my_list:
    y=str(i)
    #print(y)
    #word=nltk.word_tokenize(y)
    #print(word)
    tokens=re.findall('\w+',y)
    #print(tokens)
    words=[]
    for word in tokens:
        words.append(word.lower())
    #print(words)
    sw=nltk.corpus.stopwords.words('english')
    #print(sw)
    words_notstop=[]
    for word in words:
        if word not in sw:
            words_notstop.append(word)
            
    # print(words_notstop)
    #frequency=len(words_notstop)
    #print(frequency)
    frewords=nltk.FreqDist(words_notstop)
    #d["Frequency"]=frewords
    print(frewords)
    frewords.plot()
```
